```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
path="https://raw.githubusercontent.com/Anasuya-Sahoo/DMDW-Lab/main/student-mat.csv"
data = pd.read_csv(path)
data
```

```python
data.shape
```

```python
data.head()
```

```python
dfs = data[['sex','age','famsize','Fjob','paid','higher','reason','famsup','Mjob']]
dfs.head()
```

```python
dfs = dfs.replace('no',0)
dfs = dfs.replace('yes',1)
dfs.head()
```

```python
n=np.array(dfs[['age','paid']])
n=n.reshape(-1,2)
n.shape
```

```python
m=np.array(dfs[['higher','famsup']])
m.reshape(-1,2)
m.shape
```

```python
from scipy.spatial.distance import cdist
```

```python
dist_matrix = cdist(n,m)
print(dist_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(dist_matrix)
plt.show()
```

```python
nominal=df[['Mjob','Fjob','reason','guardian']]
nominal=nominal.replace('at_home','home')
nominal=(nominal.astype('category'))
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
nominal['guardian']=lb.fit_transform(nominal['guardian'])
nominal['Mjob']=lb.fit_transform(nominal['Mjob'])
nominal['Fjob']=lb.fit_transform(nominal['Fjob'])
nominal['reason']=lb.fit_transform(nominal['reason'])
```

```python
nominal.head()
```

```python
nom1=np.array(nominal)
nom1.reshape(-1,2)
nom2=np.array(nominal)
nom2.reshape(-1,2)
```

```python
from scipy.spatial import distance
dist_matrix=distance.cdist(nom1,nom2)
sns.heatmap(dist_matrix)
dist_matrix.shape
```
